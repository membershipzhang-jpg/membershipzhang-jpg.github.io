# Project 2 : Are Americans Happier with More Cows?

This project will look at the evolvement of relationship between the number of cattles in the United States and people's happines. (A silly research topic, yes) through a simple scatterplot visualization. This project will look at happiness scores reported in World Happiness Report and cattle counts from Food and Agriculture Organization.

## Step 1

Importing and organizing the data. Extracted from our world in data. 

```python
import pandas as pd
import plotly.express as px
from IPython.display import HTML

# Load datasets

cattle = pd.read_csv("cattle.csv")
happiness = pd.read_csv("happiness.csv")

# Rename columns
cattle = cattle.rename(
    columns={
        "Entity": "country",
        "Code": "code",
        "Year": "year",
        "Cattle | 00000866 || Stocks | 005111 || animals": "cattle_heads",
    }
)

happiness = happiness.rename(
    columns={
        "Entity": "country",
        "Code": "code",
        "Year": "year",
        "Cantril ladder score": "happiness_score",
    }
)
```

## Step 2
Data-cleaning, merge and filtering for the United States.

```python
# Sort to U.S only
cattle_us = cattle[cattle["country"] == "United States"]
happiness_us = happiness[happiness["country"] == "United States"]


# Merge datasets
merged = pd.merge(
    cattle_us[["year", "cattle_heads"]],
    happiness_us[["year", "happiness_score"]],
    on="year",
    how="inner",
)

merged.head()

# Melt
plot_df = pd.melt(
    merged,
    id_vars="year",
    value_vars=["cattle_heads", "happiness_score"],
    var_name="variable",
    value_name="value",
)
```

## Step 3
Line-Plot and Scatter-Plot

```python
# Line plot
fig = px.line(
    plot_df,
    x="year",
    y="value",
    color="variable",
    title="U.S. Cattle Population vs Happiness Over Time",
    labels={"year": "Year", "value": "Value", "variable": "Series"},
)

HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))
```

```python
# Scatterplot
fig2 = px.scatter(
    merged,
    x="cattle_heads",
    y="happiness_score",
    title="Are Americans Happier When There Are More Cattle?",
    labels={"cattle_heads": "Cattle (Heads)", "happiness_score": "Happiness Score"},
)

HTML(fig2.to_html(include_plotlyjs="cdn", full_html=False))
```

## Takeaways

As one may assume, there appears to be no association between American's happiness and cow count judging from simple scatter and line plots. Although for the line plot it is more difficult to gauge as the units for cows are so much bigger than the units for happines score, which fluctuates within 1.0 interval over years. If we learn more coding methods to smooth the units then we can find a better appearance for this graph. For now, America's happiness does not depend on its cows. Even if we observe similar trends, correlation does not equal to causation.
